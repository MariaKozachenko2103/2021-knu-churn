```python
import pandas as pd
import numpy as np
import math as m
import random as rd
from scipy import stats
from sklearn.preprocessing import LabelEncoder

def find_missing_values():
    missing_vals = []
    num_cols = len(data.columns)
    for i in range(num_cols):
        if data.isnull().sum()[i] != 0:
            missing_vals.append(data.columns[i])
    return missing_vals

def input_missing_values(mis_vals):
    if mis_vals != []:
        for col in mis_vals:
            data[col].mean()

def find_incorrect_values():
    incorrect_values = []
    for col in data.columns:
        for val in data[col]:
            if val == "Unknown":
                incorrect_values.append(col)
                break
    return incorrect_values

def change_incorrect_values(inc_vals):
    if inc_vals != []:
        for val in inc_vals:
            change_unknown(val)

def change_unknown(col):
    pcts_dict, unknown_num = calc_pcts(col)
    unknown_people = {}
    for name,pct in pcts_dict.items():
        unknown_people[name] = round(unknown_num*pct)
    new_col = []
    for val in data[col]:
        if val == "Unknown":
            ran_name = rd.randint(1, len(unknown_people))
            name = list(unknown_people.keys())[ran_name-1]
            while True:
                if unknown_people[name] != 0:
                    unknown_people[name] -= 1
                    break
                else:
                    ran_name = rd.randint(1, len(unknown_people))
                    name = list(unknown_people.keys())[ran_name-1]
            new_col.append(name)
        else:
            new_col.append(val)
    data[col] = new_col
            
def calc_pcts(col):
    slices = []
    labels = list(set(data[col]))
    for i in range(len(labels)):
        if labels[i] == "Unknown":
            labels.pop(i)
            break
    for sector in labels:
        slices.append(list(data[col]).count(sector))
    pcts_list = []
    sum = np.sum(slices)
    for val in slices:
        pcts_list.append(val/sum) 
    pcts_dict = {}
    for i in range(len(labels)):
        pcts_dict[labels[i]] = pcts_list[i]
    unknown_num = list(data[col]).count("Unknown")
    return (pcts_dict, unknown_num)

def step1():
    input_missing_values(find_missing_values())
    change_incorrect_values(find_incorrect_values())

def str_to_int(col, str_vals, int_vals):
    new_col = []
    for val in data[col]:
        for i in range(len(str_vals)):
            if str_vals[i] == val:
                new_col.append(int_vals[i])
    data[col] = new_col
    
def sti_Attr_Flag():
    name = "Attrition_Flag"
    str_vals = ['Attrited Customer', 'Existing Customer']
    int_vals = [ 1, 0 ]
    return (name, str_vals, int_vals)

def sti_Gender():
    name = "Gender"
    str_vals = ['M', 'F']
    int_vals = [ 1, 0 ]
    return (name, str_vals, int_vals)

def sti_Mar_Stat():
    name = "Marital_Status"
    str_vals = ['Single', 'Divorced', 'Married']
    int_vals = [ 0, 1, 2 ]
    return (name, str_vals, int_vals)

def sti_Card_Cat():
    name = "Card_Category"
    str_vals = ['Blue', 'Silver', 'Gold', 'Platinum']
    int_vals = [ 0, 1, 2, 3 ]
    return (name, str_vals, int_vals)

def sti_Educ_Lvl():
    name = "Education_Level"
    str_vals = ['Uneducated', 'High School', 'College', 'Graduate', 'Post-Graduate', 'Doctorate']
    int_vals = [ 0, 1, 2, 3, 4, 5 ]
    return (name, str_vals, int_vals)

def sti_Income_Cat():
    name = "Income_Category"
    str_vals = ['Less than $40K', '$40K - $60K', '$60K - $80K', '$80K - $120K', '$120K +']
    int_vals = [ 20000, 50000, 70000, 100000, 140000 ]
    return (name, str_vals, int_vals)

def step3():
    #n - name, s - str_vals, i - int_vals
    n,s,i = sti_Attr_Flag()
    str_to_int(n,s,i)
    n,s,i = sti_Gender()
    str_to_int(n,s,i)
    n,s,i = sti_Mar_Stat()
    str_to_int(n,s,i)
    n,s,i = sti_Card_Cat()
    str_to_int(n,s,i)
    n,s,i = sti_Educ_Lvl()
    str_to_int(n,s,i)
    n,s,i = sti_Income_Cat()
    str_to_int(n,s,i)

data = pd.read_csv("BankChurners.csv")
    
print("\nOLD DATA")
print("====================================================")
print(data)
print("====================================================")
step1()
step3()
data = data[(np.abs(stats.zscore(data)) < 3).all(axis=1)]    #step2:  discovering outliers
print("\n\nNEW DATA")
print("====================================================")
print(data)
print("====================================================")
```
