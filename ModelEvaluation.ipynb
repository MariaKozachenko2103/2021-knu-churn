```python
import pandas
import numpy as np
import seaborn as sns

from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap

from sklearn.model_selection import train_test_split 
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier  
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 
from sklearn import tree
from sklearn import neighbors

Churners = pandas.read_csv("BankChurnersUpdated.csv")
Churners
```

```python
X = Churners.iloc[:, 6:19].values 
y = Churners.iloc[:, 0].values 
X_train, X_test, y_train, y_test = train_test_split( 
              X, y, test_size = 0.3)

print(X)
print(y)
```

```python
classifier1 = tree.DecisionTreeClassifier(max_depth=5) 
classifier1 = classifier1.fit(X_train,y_train) 
predicted1 = classifier1.predict(X_test) 

print ('Confusion Matrix :') 
print(confusion_matrix(y_test, predicted1)) 
print ('Accuracy Score :',accuracy_score(y_test, predicted1)) 
print ('Report : ') 
print (classification_report(y_test, predicted1)) 
```

```python
classifier2 =  RandomForestClassifier()
classifier2 = classifier2.fit(X_train, y_train)
predicted2 = classifier2.predict(X_test)

print ('Confusion Matrix :') 
print(confusion_matrix(y_test, predicted2)) 
print ('Accuracy Score :',accuracy_score(y_test, predicted2)) 
print ('Report : ') 
print (classification_report(y_test, predicted2)) 
```

```python
n_neighbors = 15


classifier3 = neighbors.KNeighborsClassifier(n_neighbors)
classifier3.fit(X, y)
predicted3 = classifier3.predict(X_test)

print ('Confusion Matrix :') 
print(confusion_matrix(y_test, predicted3)) 
print ('Accuracy Score :',accuracy_score(y_test, predicted3)) 
print ('Report : ') 
print (classification_report(y_test, predicted3)) 
```

```python
kfold = KFold(9, shuffle=True, random_state=1)

classifier = RandomForestClassifier()
scores = []

for i in range(9):
    result = next(kfold.split(Churners), None)
    
    X_train = Churners.iloc[result[0], 6:19]
    X_test = Churners.iloc[result[1], 6:19]
    
    y_train = Churners.iloc[result[0], 0]
    y_test = Churners.iloc[result[1], 0]
    
    classifier = classifier.fit(X_train,y_train)
    predicted = classifier.predict(X_test)
    scores.append(accuracy_score(y_test, predicted))
    
    print ('Confusion Matrix :') 
    print(confusion_matrix(y_test, predicted)) 
    print ('Accuracy Score :',accuracy_score(y_test, predicted)) 
    print ('Report : ') 
    print (classification_report(y_test, predicted)) 
    

print('Average K-Fold Score :' , np.mean(scores))
```
